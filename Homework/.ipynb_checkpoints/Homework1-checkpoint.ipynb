Write a program to plot$$
f(x)=\pi/2+2*(sin(x)+sin(3x)/3+sin(5x)/5+....sin(nx)/n)
$$from -$\pi$ to $\pi$. Choose n=11, 21 and 31 and plot three curves in one figure. Please use numpy library and matplotlib.

Write your own Vpython program to generate a simple animation.

```python
# import numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt

# define a function to get the square wave
def fourierSquareWave(n, x):
    # inital offset value of pi/2
    f = np.pi/2 
    
    # for loop over ever odd element
    for i in range(1, n+1, 2):
        # add new terms to our function
        f = f + (2 * np.sin(i*x)/i)
    return f

# define out start end and points
start, end, points = -np.pi, np.pi, 1000

# define where we want the series to end
series = [11, 21 ,31]

# create x points
x = np.linspace(start, end, points)

# define figure size
plt.figure(figsize = (20, 10))

# for every element in series create a new 
# graph of that function in range
for i in range(len(series)):
    y = fourierSquareWave(series[i], x)
    plt.plot(x,y, label = 'n =' + str(series[i]))
    
# plotting the functions
font = 20
plt.title('Fourier Series Approximation of Square Wave', size = font)
plt.ylabel('f(x)', size = font)
plt.xlabel('x', size = font)
plt.legend(fontsize = font)
plt.show()
```
