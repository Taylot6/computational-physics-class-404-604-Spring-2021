```python
import numpy as np
import matplotlib.pyplot as plt

n = 5
x = 0
h = 1e-4
ymin = 1.5
ymax = 2.5
maxIter = 100
target = 1
goal = 1e-4

def f(x, y):
    f=np.zeros(2)
    f[0]=y[1]
    f[1]=(1-x**2)**(-1) * (2*x*f[0] - n*(n+1)*y[0])
    return f

def rk4(x, h, y, n):
    k1 = np.zeros(n)
    k2 = np.zeros(n)
    k3 = np.zeros(n)
    k4 = np.zeros(n)
    k1=h*f(x,y)
    k2=h*f(x+0.5*h,y+0.5*k1)
    k3=h*f(x+0.5*h,y+0.5*k2)
    k4=h*f(x+h,y+k3)
    y+=(1/6) * (k1 + 2*k2 + 2*k3 + k4)
    return y

xp = []
yp = []
plt.figure(figsize=(10, 10))

for iter in range(maxIter):
    exitFlag = False
    xp=[]
    yp=[]
    yGuess = (ymin+ymax)/2
    print('Iteration',iter, 'Guess',yGuess)
    y = [0, yGuess]
    x = 0
    while(x <= 1 - 1 * h):
        y = rk4(x,h,y,2)
        yp = np.append(yp,y[0])
        xp = np.append(xp,x)
        x = x+h
    print('Current Value =', y[0])
    if (np.abs(y[0] - target) < goal):
        print('Final Answer, y_a = ', yGuess,
              'value at x',target, '=', y[0])
        exitFlag = True
    if (y[0] > target):
        ymax = yGuess
    else:
        ymin = yGuess
 
        

    plt.title('Shooting Method')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.plot(xp,yp, label = yGuess)
    if exitFlag == True:
        break

plt.legend()
plt.show()
```
