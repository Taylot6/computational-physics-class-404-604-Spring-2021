```python
import numpy as np
import matplotlib.pyplot as plt

g = 9.8
t = 0
tend = 5
h=0.0001
y=[0.,10.]

def f(t, y):
    f=np.zeros(2)
    f[0]=y[1]
    f[1]=-g
    return f

def rk4(t, h, y, n):
    k1 = np.zeros(n)
    k2 = np.zeros(n)
    k3 = np.zeros(n)
    k4 = np.zeros(n)
    k1=h*f(t,y)
    k2=h*f(t+0.5*h,y+0.5*k1)
    k3=h*f(t+0.5*h,y+0.5*k2)
    k4=h*f(t+h,y+k3)
    y+=(1/6) * (k1 + 2*k2 + 2*k3 + k4)
    return y

xp=[]
tp=[]
while(y[0]>=0):
    y = rk4(t,h,y,4)
    xp=np.append(xp,y[0])
    tp=np.append(tp,t)
    t = t+h

plt.title('Trown Ball Height as a Function of Time')
plt.xlabel('time (s)')
plt.ylabel('height (m)')
plt.plot(tp,xp)
plt.show()

def analyticSolution(t):
    return -0.5*9.8*t**2 + 10*t

plt.plot(tp, analyticSolution(tp),label='Analytic')
plt.title('Analytic and Numerical solutions')
plt.xlabel('time (s)')
plt.ylabel('height (m)')
plt.plot(tp,xp,label='RungeKutta')
plt.legend()

plt.show()

plt.plot(tp, analyticSolution(tp)-xp, label= 'Difference')
plt.title('Difference Between Analytic and Numerical (Residual)')
plt.xlabel('time (s)')
plt.ylabel('height (m)')
plt.legend()

plt.show()
```
